```python
batsman_stats = deliveries.groupby('batter').agg({
    'batsman_runs': ['sum', 'mean', 'count'],
    'is_wicket': 'sum'
}).round(2)
batsman_stats.columns = ['total_runs', 'avg_runs_per_ball', 'balls_faced', 'times_out'] # Total run , Average run per ball , Number of ball faced , wicket goten
batsman_stats['strike_rate'] = (batsman_stats['total_runs'] / batsman_stats['balls_faced'] * 100).round(2) # Strike Rate = (Runs / Balls) × 100

batsman_stats['matches_played'] = (batsman_stats['balls_faced'] / 120).round(0)
batsman_stats['avg_runs_per_match'] = (batsman_stats['total_runs'] / batsman_stats['matches_played']).round(2)# Avg runs per match (assuming 6 balls per over = 1 match ~ 120 balls)

# Consistency (standard deviation of runs per ball)
batsman_consistency = deliveries.groupby('batter')['batsman_runs'].std().round(2)
batsman_stats['consistency'] = batsman_consistency

batsman_stats = batsman_stats.sort_values('total_runs', ascending=False)

print(batsman_stats.head(15))
```

```python
# Create bowler statistics
bowler_stats = deliveries.groupby('bowler').agg({
    'total_runs': 'sum',
    'is_wicket': 'sum'
}).round(2)

bowler_stats.columns = ['runs_given', 'wickets']
bowler_stats['balls_bowled'] = deliveries.groupby('bowler').size()

# Economy Rate = (Runs Given / Overs) = (Runs / (Balls / 6))
bowler_stats['economy_rate'] = (bowler_stats['runs_given'] / (bowler_stats['balls_bowled'] / 6)).round(2)

# Wickets per match (assuming 60 balls = 1 match for a bowler)
bowler_stats['matches_bowled'] = (bowler_stats['balls_bowled'] / 60).round(0)
bowler_stats['wickets_per_match'] = (bowler_stats['wickets'] / bowler_stats['matches_bowled']).round(2)

# Efficiency = Wickets / Runs Given
bowler_stats['efficiency'] = (bowler_stats['wickets'] / bowler_stats['runs_given']).round(3)

bowler_stats = bowler_stats.sort_values('wickets', ascending=False)

print(bowler_stats.head(15))
```

```python
# Team wins
team_wins = matches['winner'].value_counts().reset_index()
team_wins.columns = ['team', 'total_wins']

# Total matches for each team
team_matches_1 = matches['team1'].value_counts().reset_index()
team_matches_1.columns = ['team', 'matches_as_team1']

team_matches_2 = matches['team2'].value_counts().reset_index()
team_matches_2.columns = ['team', 'matches_as_team2']

# Merge all
team_features = team_wins.copy()
team_features = team_features.merge(team_matches_1, on='team', how='left').fillna(0)
team_features = team_features.merge(team_matches_2, on='team', how='left').fillna(0)

team_features['total_matches'] = team_features['matches_as_team1'] + team_features['matches_as_team2']

# Win Rate = Wins / Total Matches × 100
team_features['win_rate'] = (team_features['total_wins'] / team_features['total_matches'] * 100).round(2)

team_features = team_features.sort_values('win_rate', ascending=False)

print(team_features)
```

```python
# Wins at home (team1 position = home)
home_wins = matches[matches['team1'] == matches['winner']].groupby('team1').size().reset_index()
home_wins.columns = ['team', 'home_wins']

# Total home matches (team1 position = home)
home_matches = matches.groupby('team1').size().reset_index()
home_matches.columns = ['team', 'home_matches']

# Merge
home_advantage = home_matches.merge(home_wins, on='team', how='left').fillna(0)

# Home Win Rate
home_advantage['home_win_rate'] = (home_advantage['home_wins'] / home_advantage['home_matches'] * 100).round(2)

print("✅ Home Ground Advantage Features Created!")
print(home_advantage.sort_values('home_win_rate', ascending=False))
```

```python
# First, merge deliveries with matches to get season
deliveries_with_season = deliveries.merge(
    matches[['id', 'season']],
    left_on='match_id',
    right_on='id',
    how='left'
)

# Batsman performance per season
batsman_season_stats = deliveries_with_season.groupby(['season', 'batter'])['batsman_runs'].agg(['sum', 'mean', 'count']).reset_index()
batsman_season_stats.columns = ['season', 'batter', 'runs_in_season', 'avg_per_ball', 'balls_in_season']

# Consistency across seasons (std dev of season runs)
batsman_consistency_across_seasons = batsman_season_stats.groupby('batter')['runs_in_season'].agg(['mean', 'std', 'count']).round(2)
batsman_consistency_across_seasons.columns = ['avg_runs_per_season', 'season_consistency', 'seasons_played']

print("✅ Season Consistency Features Created!")
print(batsman_consistency_across_seasons.head(15))
```

```python
# Average runs per over (all matches)
runs_per_over = deliveries.groupby('over')['total_runs'].agg(['sum', 'mean', 'count']).round(2)
runs_per_over.columns = ['total_runs', 'avg_runs', 'deliveries_count']

# Powerplay (first 6 overs)
powerplay_avg = deliveries[deliveries['over'] < 6]['total_runs'].mean()

# Middle overs (6-15)
middle_avg = deliveries[(deliveries['over'] >= 6) & (deliveries['over'] < 16)]['total_runs'].mean()

# Death overs (16+)
death_avg = deliveries[deliveries['over'] >= 16]['total_runs'].mean()

print("✅ Match Phase Features Created!")
print(f"Powerplay (0-5 overs)   Avg Runs: {powerplay_avg:.2f}")
print(f"Middle Overs (6-15)     Avg Runs: {middle_avg:.2f}")
print(f"Death Overs (16+)       Avg Runs: {death_avg:.2f}")
```

```python
# How batsmen get out
dismissal_types = deliveries[deliveries['is_wicket'] == 1]['dismissal_kind'].value_counts()

batsman_dismissals = deliveries[deliveries['is_wicket'] == 1].groupby('batter')['dismissal_kind'].value_counts().unstack(fill_value=0)

print("✅ Dismissal Pattern Features Created!")
print("\nDismissal Types Across IPL:")
print(dismissal_types)
print("\nTop 10 Batsmen Dismissal Patterns:")
print(batsman_dismissals.head(10))
```

```python
# Create match_id to venue mapping
match_venue_map = matches[['id', 'venue']].rename(columns={'id': 'match_id'})

# Merge venue info with deliveries
deliveries_with_venue = deliveries.merge(match_venue_map, on='match_id', how='left')

# Matches at each venue
venue_matches = matches['venue'].value_counts().reset_index()
venue_matches.columns = ['venue', 'total_matches']

# Average runs scored at each venue
venue_avg_runs = deliveries_with_venue.groupby('venue')['total_runs'].mean().reset_index()
venue_avg_runs.columns = ['venue', 'avg_runs_per_match']

# Merge
venue_features = venue_matches.merge(venue_avg_runs, on='venue', how='left')
venue_features = venue_features.sort_values('avg_runs_per_match', ascending=False)

print("✅ Venue Features Created!")
print(venue_features.head(15))
```

```python
# Most valuable players (Player of Match awards)
pom_awards = matches[matches['player_of_match'] != 'Unknown']['player_of_match'].value_counts().reset_index()
pom_awards.columns = ['player', 'pom_awards']

# Total matches played by each player
player_total_matches = deliveries.groupby('batter')['match_id'].nunique().reset_index()
player_total_matches.columns = ['player', 'total_matches']

# Merge
pom_stats = pom_awards.merge(player_total_matches, left_on='player', right_on='player', how='left').fillna(0)

# Percentage of matches where player got PoM
pom_stats['pom_percentage'] = (pom_stats['pom_awards'] / pom_stats['total_matches'] * 100).round(2)

print("✅ Player of Match Features Created!")
print(pom_stats.head(15))
```
